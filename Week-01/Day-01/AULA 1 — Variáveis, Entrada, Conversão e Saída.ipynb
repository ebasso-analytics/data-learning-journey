Crie um programa que receba nome, idade e cidade, e exiba a mensagem formatada desta forma:

Regras:

idade deve ser um n√∫mero inteiro

EXERC√çCIO 2 ‚Äî Classifica√ß√£o de Nota (Condicionais)

Objetivo: praticar entrada de dados, convers√£o, valida√ß√£o e uso de if/elif/else.

Enunciado:

Solicite ao usu√°rio uma nota de 0 a 10 e retorne o conceito conforme a tabela:

Nota	Conceito
9.0 a 10	A
8.0 a 8.9	B
7.0 a 7.9	C
6.0 a 6.9	D
abaixo de 6.0	F

Regras:
‚úî Validar se o usu√°rio digitou n√∫mero
‚úî Repetir a pergunta enquanto for inv√°lido
‚úî Mostrar mensagem final: "Sua nota √© X e seu conceito √© Y"

Dica: pode usar while True e try/except.

```python
# Valida√ß√£o da nota digitada 
while True:
    nota = input("Digite a sua nota (0 a 10): ").strip()
    try:
        nota = float(nota)
        if 0 <= nota <= 10:
            break
        else:
            print("A nota deve estar entre 0 e 10. Tente novamente!")
    except ValueError:
        print("Valor inv√°lido! Digite apenas n√∫meros (ex: 7, 8.5, 10).")

# Classifica√ß√£o da nota
if nota >= 9:
    conceito = "A"
elif nota >= 8:
    conceito = "B"
elif nota >= 7:
    conceito = "C"
elif nota >= 6:
    conceito = "D"
else:
    conceito = "F"

print(f"Sua nota √© {nota} e seu conceito √© {conceito}.")
```

üìò Exerc√≠cio 3

Crie um programa que pe√ßa ao usu√°rio um n√∫mero inteiro e verifique se ele √©:

Par ou √≠mpar

Positivo, negativo ou zero

Tamb√©m deve informar se o n√∫mero √© maior, menor ou igual a 100

üìç Regras obrigat√≥rias:

O programa s√≥ deve aceitar n√∫meros inteiros

Usar valida√ß√£o com while True

A sa√≠da deve ser amig√°vel e clara ao usu√°rio

Utilizar m√≠nimo de repeti√ß√µes poss√≠veis no c√≥digo

```python


while True:
    numero_inteiro = input("Digite um n√∫mero inteiro de 0 a 100: ").strip()
    try:
        numero_inteiro = int(numero_inteiro)
    except ValueError:
        print("Valor inv√°lido! Digite apenas um n√∫mero inteiro de -100 a 100! ")
        continue
    
    if -100 <= numero_inteiro <= 100:
        break
    else:
        print("Digite apenas um n√∫mero inteiro de -100 a 100. Tente novamente!")

#Verifica√ß√µes:   
# Par ou √≠mpar     
if numero_inteiro % 2 == 0:
    print("O n√∫mero √© PAR.")
else:
    print("O n√∫mero √© IMPAR.")

# Positivo, negativo ou zero
if numero_inteiro > 0:
    print("O n√∫mero √© POSITIVO.")
elif numero_inteiro < 0:
    print("O n√∫mero √© NEGATIVO.")
else:
    print("O n√∫mero √© ZERO.")

# Compara√ß√£o com 100
if numero_inteiro > 100:
    print("Maior que 100.")
elif numero_inteiro < 100:
    print("Menor que 100.")
else:
    print("Igual a 100.")
```

‚úÖ EXERC√çCIO 4 ‚Äî Listas, loops e estat√≠sticas b√°sicas

Voc√™ vai criar um programa que:

1Ô∏è‚É£ Recebe do usu√°rio uma lista de n√∫meros

O usu√°rio digita v√°rios n√∫meros separados por v√≠rgula, por exemplo:
10, 5, 22, 3, 7

2Ô∏è‚É£ Seu programa deve:

Converter essa entrada em uma lista de inteiros

Validar se todos s√£o n√∫meros

Calcular:

Soma dos valores

M√©dia

Maior valor

Menor valor

Quantidade de n√∫meros

Exibir tudo de forma organizada

3Ô∏è‚É£ Exemplo de sa√≠da esperada:
Lista recebida: [10, 5, 22, 3, 7]
Quantidade de n√∫meros: 5
Soma: 47
M√©dia: 9.4
Maior valor: 22
Menor valor: 3

üéØ Regras

N√ÉO pode usar: sum(), max(), min() (vamos treinar l√≥gica pura).

Deve usar:

for

Convers√£o de tipos

Listas

Acumula√ß√£o manual

üß† Dicas para montar seu c√≥digo

Receba a entrada como string

Divida usando .split(',')

Valide cada item

Converta para int

V√° acumulando soma, maior e menor com um for

Depois calcule m√©dia = soma / quantidade

```python
# Entrada de n√∫meros para gera√ß√£o de lista

while True:
    entrada = input("Digite v√°rios n√∫meros separados por v√≠rgula: ").strip()

    # Divide a string em partes
    partes = entrada.split(",")

    lista_numeros = []
    valido = True  # Flag para verificar se todos s√£o n√∫meros v√°lidos

    # Valida√ß√£o e convers√£o manual
    for item in partes:
        item = item.strip()  # Remove espa√ßos

        if item.replace("-", "").isdigit():  # Permite n√∫meros negativos
            lista_numeros.append(int(item))
        else:
            print(f"Valor inv√°lido encontrado: '{item}'. Tente novamente.\n")
            valido = False
            break

    if valido:
        break  # Sai do loop se estiver tudo ok

    # calcular tudo manualmente

quantidade = 0
soma = 0

# maior e menor com o primeiro elemento
maior = lista_numeros[0]
menor = lista_numeros[0]

# Percorre a lista com for
for num in lista_numeros:
    quantidade += 1
    soma += num

    if num > maior:
        maior = num

    if num < menor:
        menor = num

# C√°lculo da m√©dia
media = soma / quantidade

# Exibi√ß√£o final
print("\n=== RESULTADOS ===")
print(f"Lista recebida: {lista_numeros}")
print(f"Quantidade de n√∫meros: {quantidade}")
print(f"Soma: {soma}")
print(f"M√©dia: {media}")
print(f"Maior valor: {maior}")
print(f"Menor valor: {menor}")
```

Este c√≥digo faz exatamente o mesmo exerc√≠cio, mas agora dividido em fun√ß√µes profissionais.

‚úî Como quebrar um problema em fun√ß√µes
‚úî Reutiliza√ß√£o de c√≥digo
‚úî Organiza√ß√£o de l√≥gica
‚úî Fun√ß√µes retornando m√∫ltiplos valores
‚úî C√≥digo profissional e limpo

```python
def ler_lista_numeros():
    """L√™ a entrada do usu√°rio e retorna uma lista de inteiros v√°lida."""
    while True:
        entrada = input("Digite v√°rios n√∫meros separados por v√≠rgula: ").strip()

        partes = entrada.split(",")
        lista = []
        valido = True

        for item in partes:
            item = item.strip()

            # Valida√ß√£o manual (sem try, por enquanto)
            if item.replace("-", "").isdigit():
                lista.append(int(item))
            else:
                print(f"Valor inv√°lido encontrado: '{item}'. Tente novamente.\n")
                valido = False
                break
        
        if valido:
            return lista


def calcular_estatisticas(lista):
    """Recebe uma lista e calcula soma, m√©dia, maior, menor e quantidade manualmente."""
    quantidade = 0
    soma = 0
    maior = lista[0]
    menor = lista[0]

    for num in lista:
        quantidade += 1
        soma += num

        if num > maior:
            maior = num
        if num < menor:
            menor = num

    media = soma / quantidade

    return quantidade, soma, media, maior, menor


def exibir_resultados(lista, quantidade, soma, media, maior, menor):
    """Mostra os resultados de forma organizada."""
    print("\n=== RESULTADOS ===")
    print(f"Lista recebida: {lista}")
    print(f"Quantidade de n√∫meros: {quantidade}")
    print(f"Soma: {soma}")
    print(f"M√©dia: {media}")
    print(f"Maior valor: {maior}")
    print(f"Menor valor: {menor}")


# --- PROGRAMA PRINCIPAL ---
lista_numeros = ler_lista_numeros()
quantidade, soma, media, maior, menor = calcular_estatisticas(lista_numeros)
exibir_resultados(lista_numeros, quantidade, soma, media, maior, menor)
```
