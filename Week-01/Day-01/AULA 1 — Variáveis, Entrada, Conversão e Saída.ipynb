EXERC√çCIO 2 ‚Äî Classifica√ß√£o de Nota (Condicionais)

Objetivo: praticar entrada de dados, convers√£o, valida√ß√£o e uso de if/elif/else.

Enunciado:

Solicite ao usu√°rio uma nota de 0 a 10 e retorne o conceito conforme a tabela:

Nota	Conceito
9.0 a 10	A
8.0 a 8.9	B
7.0 a 7.9	C
6.0 a 6.9	D
abaixo de 6.0	F

Regras:
‚úî Validar se o usu√°rio digitou n√∫mero
‚úî Repetir a pergunta enquanto for inv√°lido
‚úî Mostrar mensagem final: "Sua nota √© X e seu conceito √© Y"

```python
# Valida√ß√£o da nota digitada 
while True:
    nota = input("Digite a sua nota (0 a 10): ").strip()
    try:
        nota = float(nota)
        if 0 <= nota <= 10:
            break
        else:
            print("A nota deve estar entre 0 e 10. Tente novamente!")
    except ValueError:
        print("Valor inv√°lido! Digite apenas n√∫meros (ex: 7, 8.5, 10).")

# Classifica√ß√£o da nota
if nota >= 9:
    conceito = "A"
elif nota >= 8:
    conceito = "B"
elif nota >= 7:
    conceito = "C"
elif nota >= 6:
    conceito = "D"
else:
    conceito = "F"

print(f"Sua nota √© {nota} e seu conceito √© {conceito}.")
```

‚úÖ EXERC√çCIO 4 ‚Äî Listas, loops e estat√≠sticas b√°sicas

O usu√°rio digita v√°rios n√∫meros separados por v√≠rgula, por exemplo:

Calcular:

V√° acumulando soma, maior e menor com um for

```python
# Entrada de n√∫meros para gera√ß√£o de lista

while True:
    entrada = input("Digite v√°rios n√∫meros separados por v√≠rgula: ").strip()

    # Divide a string em partes
    partes = entrada.split(",")

    lista_numeros = []
    valido = True  # Flag para verificar se todos s√£o n√∫meros v√°lidos

    # Valida√ß√£o e convers√£o manual
    for item in partes:
        item = item.strip()  # Remove espa√ßos

        if item.replace("-", "").isdigit():  # Permite n√∫meros negativos
            lista_numeros.append(int(item))
        else:
            print(f"Valor inv√°lido encontrado: '{item}'. Tente novamente.\n")
            valido = False
            break

    if valido:
        break  # Sai do loop se estiver tudo ok

    # calcular tudo manualmente

quantidade = 0
soma = 0

# maior e menor com o primeiro elemento
maior = lista_numeros[0]
menor = lista_numeros[0]

# Percorre a lista com for
for num in lista_numeros:
    quantidade += 1
    soma += num

    if num > maior:
        maior = num

    if num < menor:
        menor = num

# C√°lculo da m√©dia
media = soma / quantidade

# Exibi√ß√£o final
print("\n=== RESULTADOS ===")
print(f"Lista recebida: {lista_numeros}")
print(f"Quantidade de n√∫meros: {quantidade}")
print(f"Soma: {soma}")
print(f"M√©dia: {media}")
print(f"Maior valor: {maior}")
print(f"Menor valor: {menor}")
```

Este c√≥digo faz exatamente o mesmo exerc√≠cio, mas agora dividido em fun√ß√µes profissionais.

```python
def ler_lista_numeros():
    """L√™ a entrada do usu√°rio e retorna uma lista de inteiros v√°lida."""
    while True:
        entrada = input("Digite v√°rios n√∫meros separados por v√≠rgula: ").strip()

        partes = entrada.split(",")
        lista = []
        valido = True

        for item in partes:
            item = item.strip()

            # Valida√ß√£o manual (sem try, por enquanto)
            if item.replace("-", "").isdigit():
                lista.append(int(item))
            else:
                print(f"Valor inv√°lido encontrado: '{item}'. Tente novamente.\n")
                valido = False
                break
        
        if valido:
            return lista


def calcular_estatisticas(lista):
    """Recebe uma lista e calcula soma, m√©dia, maior, menor e quantidade manualmente."""
    quantidade = 0
    soma = 0
    maior = lista[0]
    menor = lista[0]

    for num in lista:
        quantidade += 1
        soma += num

        if num > maior:
            maior = num
        if num < menor:
            menor = num

    media = soma / quantidade

    return quantidade, soma, media, maior, menor


def exibir_resultados(lista, quantidade, soma, media, maior, menor):
    """Mostra os resultados de forma organizada."""
    print("\n=== RESULTADOS ===")
    print(f"Lista recebida: {lista}")
    print(f"Quantidade de n√∫meros: {quantidade}")
    print(f"Soma: {soma}")
    print(f"M√©dia: {media}")
    print(f"Maior valor: {maior}")
    print(f"Menor valor: {menor}")


# --- PROGRAMA PRINCIPAL ---
lista_numeros = ler_lista_numeros()
quantidade, soma, media, maior, menor = calcular_estatisticas(lista_numeros)
exibir_resultados(lista_numeros, quantidade, soma, media, maior, menor)
```

```python
def coletar_lista():
    """Solicita ao usu√°rio n√∫meros separados por v√≠rgula e devolve uma lista de inteiros validada."""
    while True:
        entrada = input("Digite v√°rios n√∫meros separados por v√≠rgula: ").strip()
        partes = entrada.split(",")
        lista = []
        valido = True

        for item in partes:
            item = item.strip()

            # Verifica√ß√£o manual para n√∫meros e negativos
            if item.startswith("-"):
                if item[1:].isdigit():
                    lista.append(int(item))
                else:
                    print(f"Valor inv√°lido encontrado: '{item}'. Tente novamente!")
                    valido = False
                    break
            else:
                if item.isdigit():
                    lista.append(int(item))
                else:
                    print(f"Valor inv√°lido encontrado: '{item}'. Tente novamente!")
                    valido = False
                    break

        if valido:
            return lista


def calcular_estatisticas(lista):
    """Recebe uma lista de inteiros e retorna um dicion√°rio com soma, m√©dia, maior, menor e quantidade."""
    soma = 0
    quantidade = 0
    maior = lista[0]
    menor = lista[0]

    for num in lista:
        soma += num
        quantidade += 1

        if num > maior:
            maior = num
        if num < menor:
            menor = num

    media = soma / quantidade

    return {
        "soma": soma,
        "media": media,
        "maior": maior,
        "menor": menor,
        "quantidade": quantidade
    }


def exibir_resultados(lista, stats):
    """Exibe os resultados de forma organizada."""
    print("\n=== RESULTADOS ===")
    print(f"Lista recebida: {lista}")
    print(f"Quantidade de n√∫meros: {stats['quantidade']}")
    print(f"Soma: {stats['soma']}")
    print(f"M√©dia: {stats['media']}")
    print(f"Maior valor: {stats['maior']}")
    print(f"Menor valor: {stats['menor']}")


# --------------- PROGRAMA PRINCIPAL ---------------
lista = coletar_lista()
estatisticas = calcular_estatisticas(lista)
exibir_resultados(lista, estatisticas)
```

pode usar esse padr√£o para:

```python
def coletar_lista():
    while True:
        entrada = input("Digite n√∫meros separados por v√≠rgula: ").strip()

        # Divide, limpa espa√ßos e forma uma lista
        partes = [p.strip() for p in entrada.split(",")]

        lista = []
        valido = True

        for item in partes:

            # Tratamento para n√∫meros negativos
            if item.startswith("-") and item[1:].isdigit():
                lista.append(int(item))
            elif item.isdigit():
                lista.append(int(item))
            else:
                print(f"Valor inv√°lido encontrado: '{item}'. Tente novamente!")
                valido = False
                break

        if valido:
            return lista


def calcular_estatisticas(lista):
    soma = 0
    maior = lista[0]
    menor = lista[0]
    qtd = 0

    for n in lista:
        soma += n
        qtd += 1

        if n > maior:
            maior = n
        if n < menor:
            menor = n

    media = soma / qtd

    return soma, media, maior, menor, qtd


def exibir(lista, soma, media, maior, menor, qtd):
    print("\n=== RESULTADOS (Vers√£o Avan√ßada) ===")
    print(f"Lista: {lista}")
    print(f"Quantidade: {qtd}")
    print(f"Soma: {soma}")
    print(f"M√©dia: {media}")
    print(f"Maior: {maior}")
    print(f"Menor: {menor}")


# ---------------- PROGRAMA PRINCIPAL ----------------
lista = coletar_lista()
soma, media, maior, menor, qtd = calcular_estatisticas(lista)
exibir(lista, soma, media, maior, menor, qtd)
```

F√°cil de ler, manter e evoluir.

```python
from typing import List, Dict


def parse_input(entrada: str) -> List[int]:
    """
    Converte uma string com n√∫meros separados por v√≠rgula
    para uma lista de inteiros.

    Args:
        entrada (str): Texto digitado pelo usu√°rio.

    Returns:
        List[int]: Lista de inteiros convertida.

    Raises:
        ValueError: Se algum item n√£o for n√∫mero inteiro v√°lido.
    """
    partes = [p.strip() for p in entrada.split(",")]
    numeros = []

    for item in partes:
        # Permite n√∫meros negativos
        if item.startswith("-"):
            if item[1:].isdigit():
                numeros.append(int(item))
            else:
                raise ValueError(f"Valor inv√°lido encontrado: '{item}'")
        else:
            if item.isdigit():
                numeros.append(int(item))
            else:
                raise ValueError(f"Valor inv√°lido encontrado: '{item}'")

    return numeros


def calcular_estatisticas(lista: List[int]) -> Dict[str, float]:
    """
    Calcula soma, m√©dia, maior, menor e quantidade usando l√≥gica manual.

    Args:
        lista (List[int]): Lista de inteiros.

    Returns:
        Dict[str, float]: Dicion√°rio com estat√≠sticas calculadas.
    """
    soma = 0
    maior = lista[0]
    menor = lista[0]
    qtd = 0

    for n in lista:
        soma += n
        qtd += 1

        if n > maior:
            maior = n
        if n < menor:
            menor = n

    media = soma / qtd

    return {
        "soma": soma,
        "media": media,
        "maior": maior,
        "menor": menor,
        "quantidade": qtd
    }


def exibir_resultados(lista: List[int], estat: Dict[str, float]) -> None:
    """
    Exibe as estat√≠sticas de forma organizada.

    Args:
        lista (List[int]): Lista original.
        estat (Dict[str, float]): Estat√≠sticas calculadas.
    """
    print("\n=== RESULTADOS (Vers√£o Ultra Profissional) ===")
    print(f"Lista: {lista}")
    print(f"Quantidade: {estat['quantidade']}")
    print(f"Soma: {estat['soma']}")
    print(f"M√©dia: {estat['media']}")
    print(f"Maior: {estat['maior']}")
    print(f"Menor: {estat['menor']}")


def coletar_lista_usuario() -> List[int]:
    """
    Pergunta ao usu√°rio at√© que ele forne√ßa uma lista v√°lida.

    Returns:
        List[int]: Lista v√°lida de inteiros.
    """
    while True:
        entrada = input("Digite n√∫meros separados por v√≠rgula: ").strip()
        try:
            return parse_input(entrada)
        except ValueError as erro:
            print(erro)
            print("Tente novamente!\n")


# --------------------------------------------------
# ENTRY POINT PROFISSIONAL
# --------------------------------------------------
if __name__ == "__main__":
    lista = coletar_lista_usuario()
    estatisticas = calcular_estatisticas(lista)
    exibir_resultados(lista, estatisticas)
```

Uso de @property, docstrings, encapsulamento, valida√ß√£o no construtor.

‚úî Reutiliz√°vel

Essa classe pode entrar num projeto Flask, FastAPI, ETL, notebook de Data Science, tudo.

```python
from typing import List


class ListaNumerica:
    """
    Classe para representar uma lista de n√∫meros inteiros e
    calcular estat√≠sticas b√°sicas usando apenas l√≥gica manual.
    """

    def __init__(self, entrada: str):
        """
        Construtor: recebe a string digitada pelo usu√°rio,
        converte para lista de inteiros e valida.

        Args:
            entrada (str): Ex: "10, 5, -3, 22"

        Raises:
            ValueError: Se houver valores inv√°lidos.
        """
        self.lista = self._validar_lista(entrada)

    def _validar_lista(self, entrada: str) -> List[int]:
        """
        Converte texto em lista de inteiros com valida√ß√£o completa.

        Args:
            entrada (str): Texto do usu√°rio.

        Returns:
            List[int]: Lista de inteiros.

        Raises:
            ValueError: Para itens inv√°lidos.
        """
        partes = [p.strip() for p in entrada.split(",")]
        numeros = []

        for item in partes:
            if item.startswith("-") and item[1:].isdigit():
                numeros.append(int(item))
            elif item.isdigit():
                numeros.append(int(item))
            else:
                raise ValueError(f"Valor inv√°lido encontrado: '{item}'")

        return numeros

    # ======================
    # PROPRIEDADES PYTHONICAS
    # ======================

    @property
    def quantidade(self) -> int:
        return len(self.lista)

    @property
    def soma(self) -> int:
        total = 0
        for n in self.lista:
            total += n
        return total

    @property
    def media(self) -> float:
        return self.soma / self.quantidade

    @property
    def maior(self) -> int:
        maior = self.lista[0]
        for n in self.lista:
            if n > maior:
                maior = n
        return maior

    @property
    def menor(self) -> int:
        menor = self.lista[0]
        for n in self.lista:
            if n < menor:
                menor = n
        return menor

    def __repr__(self):
        """
        Representa√ß√£o amig√°vel para debug e prints profissionais.
        """
        return f"ListaNumerica({self.lista})"


# ===============================
# PROGRAMA PRINCIPAL
# ===============================
if __name__ == "__main__":
    while True:
        entrada = input("Digite n√∫meros separados por v√≠rgula: ").strip()

        try:
            ln = ListaNumerica(entrada)
            break
        except ValueError as e:
            print(e)
            print("Tente novamente!\n")

    print("\n=== RESULTADOS (Vers√£o 7 ‚Äî POO) ===")
    print(f"Lista: {ln.lista}")
    print(f"Quantidade: {ln.quantidade}")
    print(f"Soma: {ln.soma}")
    print(f"M√©dia: {ln.media}")
    print(f"Maior: {ln.maior}")
    print(f"Menor: {ln.menor}")
```
