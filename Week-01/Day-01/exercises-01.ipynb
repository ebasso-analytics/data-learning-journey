Exercício 1 — Variáveis e tipos de dados

```python
# Declaração de variáveis básicas em Python
nome = "Eduardo"
idade = 43
altura = 1.78
estudando_python = True  # True/False (boolean)

# Exibindo valores e tipos
print("Nome:", nome, "| Tipo:", type(nome))
print("Idade:", idade, "| Tipo:", type(idade))
print("Altura:", altura, "| Tipo:", type(altura))
print("Estudando Python:", estudando_python, "| Tipo:", type(estudando_python))
```

Exercício 2 — Operações matemáticas

```python
# Operações matemáticas
soma = 10 + 5
multiplicacao = 2 * 4 * 3
divisao = 15 / 2
resto = 15 % 4

print("Soma =", soma)
print("Multiplicação =", multiplicacao)  # corrigido no próximo bloco
print("Divisão =", divisao)
print("Resto da divisão =", resto)
```

Exercício 3 — Condicionais

```python
# Entrada de dados
idade_usuario = int(input("Digite sua idade: "))

# Lógica condicional
if idade_usuario < 18:
    print("Menor de idade")
elif idade_usuario <= 59:
    print("Adulto")
else:
    print("Melhor idade")




```

Exercício 4 — Loops

```python
# Loop usando for
print("Usando FOR:")
for numero in range(1, 21):
    print(numero)

# Loop usando while
print("\nUsando WHILE:")
contador = 1
while contador <= 20:
    print(contador)
    contador += 1
```

Exercício 5 — Listas

```python
nomes = ["Ana", "Bruno", "Carlos", "Daniela", "Eduarda", "Felipe"]

# Mostrar apenas o primeiro e o último item
print("Primeiro nome:", nomes[0])
print("Último nome:", nomes[-1])

# Adicionar um nome
nomes.append("Gabriel")

# Remover um nome específico
nomes.remove("Carlos")

# Ordenar lista
nomes.sort()

print("Lista final ordenada:", nomes)
```
